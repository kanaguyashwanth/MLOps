## __Introduction to Pandas__



## Step 1: Import Pandas

Let's learn about multi-indexing in pandas.

Let's first import the libraries and load the dataset.



```python
import pandas as pd
```

## Step 2: Read a CSV File

- Read the **HousePrices.csv** file using the pd.read_csv() function:

- Use **parse_dates = True** as an argument to read the date field in the correct format since, currently, the data type of the date column is **object**

```python
df = pd.read_csv("HousePrices.csv",parse_dates = True)
```

## Step 3: Display the First Few Rows

- Display the first few rows of the DataFrame using the .head() method:


```python
df.head()
```

**Observation**

The result contains the first five rows of the DataFrame.

## Step 4: Set the Multi-Level Index

Let's create a multi-level index using city and date.


- Set a multi-level index on the DataFrame using the .set_index() method, and pass the desired columns as a list

Let's modify the original DataFrame using **inplace = True**.


```python
df.set_index(['city','date'],inplace = True)
```

## Step 5: Display the DataFrame Index

- Display the DataFrame index using the .index attribute:


```python
df.index
```

**Observation**

Here, we can see a DataFrame with multiple indexes.

## Step 6: Access Rows by the Index Label

Let's access the multi-index DataFrames. This can be done using the .loc method.

Let's choose **Seattle**.

- Pass **Seattle** as an argument to df.loc[]:


```python
df.loc['Seattle']
```

**Observation**

This returns a DataFrame that contains the values of **Seattle** only.
