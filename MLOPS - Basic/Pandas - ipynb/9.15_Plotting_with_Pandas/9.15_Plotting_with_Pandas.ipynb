## Plotting with Pandas

Plotting data is required to visualize the data in Python. Python uses the plot method in pandas to create plots. There are many types of plots available in pandas.

## Step 1: Import Pandas

- Let's learn about plotting with pandas.

- Let's first import the libraries.


```python
import pandas as pd
```

## Step 2: Load the Dataset

- Load the **HousePrices.csv** dataset using the pd.read_csv() function:


```python
df = pd.read_csv('HousePrices.csv')
```

## Step 3: Display the First Few Rows

- Display the first few rows of the DataFrame using the .head() method:


```python
df.head()
```

## Step 4: Print the Columns

To print the column attribute, we need to use the column attribute.

- Print all the columns in the DataFrame:


```python
df.columns
```

## Step 5: Line Plot of All Numerical Variables

A line plot displays the information as a series of data points.

DataFrame.plot() creates a plot of a Series or a DataFrame.

- Create a line plot of all numerical variables:


```python
df.plot()
```

**Observation**

Here, we can see the information of all the numerical columns.

## Step 6: Line Plot of a Single Variable

- Create a line plot of the **price** column using the .plot() method:


```python
df['price'].plot()
```

**Observation**

The plot displays the information of **price**.

## Step 7: Add a Legend to the Plot

- Add a legend to the plot by passing the **legend=True** parameter


```python
df['price'].plot(legend=True)
```

**Observation**

The plot shows the legend of this plot.

## Step 8: Scatter Plot

A scatter plot displays the data for two variables.

- Create a scatter plot using the plot() function with the **kind='scatter'**  parameter:


```python
df.plot(x='price',y='bedrooms',kind='scatter',legend=True)
```

**Observation**

The plot displays the relationship between the data points of the variables **price** and **bedroom**.

## Step 9: Add Title and Figure Size

- Add a **title** and **figsize** to the plot; **figsize** takes a tuple as input (x-axis, y-axis).


```python
df.plot(title='Sample Plot',figsize=(15,10))
```

**Observation**

Now, the plot is resized to (15, 10) and has the title **Sample Plot**.

## Step 10: Box Plot

A box plot visualizes the summary of statistical data.

Let's create the box plot for the **price** column.

- Create a box plot using plot() with **kind=box** as a parameter
- Add a title

```python
df['price'].plot(kind='box',title='Box Plot')
```

**Observation**

The box plot shows the mean, median, quartile and outliers of the **price** column.

## Step 11: Histogram

A histogram is a graphical representation of data points organized into user-specified ranges.

- Create a histogram using plot() with the **kind=hist** parameter
- Add a title


```python
df['price'].plot(kind='hist',title='Histogram')
```

**Observation**

We can see the histogram plotted for the **price** column.
