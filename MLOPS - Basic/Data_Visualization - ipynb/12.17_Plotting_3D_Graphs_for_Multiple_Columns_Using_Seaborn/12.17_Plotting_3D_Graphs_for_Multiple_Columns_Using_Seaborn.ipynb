##__Plotting 3D Graphs for Multiple Columns Using Seaborn__
## Step 1: Import the Required Libraries and Read the Data Set

- Import pandas, seaborn and matplotlib.pyplot libraries
- Read the `ADANIPORTS.csv` file into the Python environment
- Display the first 5 rows


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.csv',parse_dates = True)
```

```python
df.head()
```

## Step 2: Calculate and Add New Columns to the DataFrame

- Calculate the difference between the 'High' and 'Low' columns, and store it in a new column 'H-L'
- Calculate the 100-day moving average for the 'Close' column, and store it in a new column '100MA'


```python
df['H-L'] = df.High - df.Low
```

```python
df['100MA'] = df['Close'].rolling(100).mean()
```

## Step 3: Plot a 3D Scatter Plot with Seaborn

- Set the style to darkgrid for better visualization
- Create a 3D scatter plot with the DataFrame index, 'H-L' column and '100MA' column


```python
sns.set_style('darkgrid')
```

```python
# Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

**Observation:**
- The result shows a 3D graph with darkgrid selected as the style.

## Step 4: Plot a 3D Spiral Graph

- Let's look at the other graphs and methods that are used for 3D plots.


```python
import numpy as np
```

Set up the data for the graph:

```python
z1 = np.linspace(0,10,100)
x1 = np.cos(4*z1)
y1 = np.sin(4*z1)
```

- Change the style to whitegrid for better visualization

- Use the plot3D method for plotting

```python
sns.set_style('whitegrid')
ax = plt.axes(projection='3d')
ax.plot3D(x1,y1,z1)
plt.show()
```

**Observation:**

- The result shows a spiral in a 3D graph.
- The spiral size can be increased by increasing the value of x1. 

## Step 5: Create  3D Surface Using Seaborn

- Define a function to calculate z values based on x and y inputs:

```python
def return_z(x,y):
    return 50-(x**2+y**2)
```

- Set the style to 'whitegrid':



```python
sns.set_style('whitegrid')
```

- Set up the data for the 3D surface plot

```python
x1,y1 = np.linspace(-5,5,50),np.linspace(-5,5,50)
```

Create a meshgrid for x and y values using NumPy, and calculate z values using the function:

```python
x1,y1 = np.meshgrid(x1,y1)
z1 = return_z(x1,y1)
```

- Create a 3D surface plot using the x1, y1 and z1 values

- Use the `plot_surface()` method to plot the 3D surface

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(x1,y1,z1)
plt.show()
```

__Observation:__

- We can see the 3D graph for multiple columns using seaborn.
