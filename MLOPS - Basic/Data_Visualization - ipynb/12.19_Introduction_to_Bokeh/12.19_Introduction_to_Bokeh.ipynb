## Step 1: Import the Required Libraries and Load the Data Set

Let's import the necessary Bokeh libraries, `figure`, `show` and `output_notebook`, and the pandas library.



```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd
```

Then, call the `output_notebook()` function to load Bokeh into the Jupyter notebook:

```python
output_notebook()
```

- Load the `HousePrices.csv` file into a DataFrame with the 'date' column as the index and parsed as dates:

- Sort the DataFrame by index:

```python
df = pd.read_csv('HousePrices.csv',index_col=['date'],parse_dates=True)
df.sort_index(inplace = True) # Sort the DataFrame by index
```

Now, display the first 5 rows:

```python
df.head()
```

## Step 2: Plot a Line Graph Using Bokeh
- Create a line plot of the 'price' column using Bokeh

- Add a line to the graph using the index and 'price' column


```python
graph = figure( title = 'Line Plot using Bokeh')
graph.line(df.index,df['price'])
show(graph)
```

**Observation:**
- The graph has multiple options to explore data, such as zooming in, saving and moving images.
