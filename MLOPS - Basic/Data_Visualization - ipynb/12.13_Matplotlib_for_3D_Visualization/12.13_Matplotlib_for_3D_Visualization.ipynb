## __Matplotlib for 3D Visualization__
## Step 1: Import the Required Libraries and Read the Data Set

- Import the pandas and matplotlib.pyplot libraries
- Load the ADANIPORTS.csv file into a DataFrame and parse the dates
- Display the first 5 rows


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.csv',parse_dates = True)
```

```python
df.head()
```

__Observation:__

- This is the head of the data set.

## Step 2: Add New Columns to the DataFrame

- Calculate the difference between the 'High' and 'Low' columns, and store it in the 'H-L' column
- Calculate the 100-day moving average of the 'Close' column, and store it in the '100MA' column


```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

## Step 3: Plot 3D Graphs

3D Scatter Plot 1

- Create a 3D scatter plot with the DataFrame index on the x-axis, 'H-L' on the y-axis and '100MA' on the z-axis
- Set x-axis, y-axis and z-axis labels
- Display the plot


```python
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

3D Scatter Plot 2 (check if there is any correlation between the columns)
- Create another 3D scatter plot with the DataFrame index on the x-axis, 'H-L' on the y-axis and 'Volume' on the z-axis
- Set the x-axis, y-axis and z-axis labels
- Display the plot

```python
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['Volume'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')
plt.show()
```

**Observation:**
- The plot shows the correlation between 'H-L' and 'volume'.
