## __Conditional Statements in Python__
## Step 1: Conditional Logic in NumPy Arrays
- There are two methods: `.where()` and `.select()`.
- These methods are very helpful while handling large data sets.

Let's understand both methods with the help of an example.

- Import the NumPy package:

```python
import numpy as np
```

Create a 1D NumPy array with values from 0 to 9:

```python
x = np.array([i for i in range(10)])
```

Display the array using its variable name, i.e, 'x':


```python
x
```

## Step 2: Use the np.where() Function

Use `np.where()` to create a new array of string values from the existing 1D array by using a conditional logic.

- Use a conditional statement to determine whether each element in 'x' is even or odd, as shown below:

- Syntax: `np.where ( condition, true expression, false expression )`

```python
np.where(x%2==0,'Even','Odd')
```

**Observation:**
- The result shows if each element is even or odd.
- It shows 'Even' where the condition is true, and prints 'Odd' outcome if the condition is false.

## Step 3: Use the np.select() Function

We first need to create a condition list and choice list to apply this function.

- The condition list includes two parameters, one for when 'x' is less than 5 and the other for when 'x' is greater than 5.

```python
condlist = [x<5, x>5]
```

The choice list contains two parameters.

1. The first parameter is the square of 'x' if 'x' is less than 5.
2. The second parameter is the cube of 'x' if 'x' is greater than 5.

```python
choicelist = [x**2,x**3]
```

Now, the condition list and choice list are ready.

- Use it in the `np.select()` method
- Syntax: `np.select(condlist,choicelist,default=x)`
- Use `np.select()` to apply the conditions from the 'condlist' and return the
 corresponding values from the 'choicelist'; if no condition is met, return
 the original value of 'x' 

```python
np.select(condlist,choicelist,default=x)
```

**Observation:**
- The result shows that if a number is less than 5, it is squared, and if it is greater than 5, then it is cubed. 
- Note that when 'x' is 5, the value remains the same because this condition is not specified in the condition list.
